# IMPORT LIBRARIES AND READ FILE

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../dataset/mental_health.csv")
```

# EDA

```python
# Donut Chart
for col in df:
    plt.figure(figsize=(8, 8))
    data = df[col].value_counts()
    colors = sns.color_palette("Set2", len(data))
    plt.pie(data, labels=data.index, colors=colors, wedgeprops=dict(width=0.4))
    plt.title(f'{col} Distribution')
    plt.show()
```
